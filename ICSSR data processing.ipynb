```python
data["Community"]=data["Community"].fillna(method='ffill')
data["Religion"]=data["Religion"].fillna(method='ffill')
data["Res Age"]=data["Res Age"].fillna("40") #mode was 40 for the age
data["Res Gender"]=data["Res Gender"].fillna(method='ffill')
data["Res Marital Sta"]=data["Res Marital Sta"].fillna(method='ffill')
data["Res Edu Sta"]=data["Res Edu Sta"].fillna(method='ffill')
data["Value.a3.a7"]=data["Value.a3.a7"].fillna(method='ffill') #Occupation status
data["Value.a3.a8"]=data["Value.a3.a8"].fillna(method='ffill') #mode of salary
data["Value.a3.a9"]=data["Value.a3.a9"].fillna(method='ffill') #salary

data["Value.a3.q55"]=data["Value.a3.q55"].fillna("3") #Place of residence number
data["Value.a4.a1"]=data["Value.a4.a1"].fillna("Rural") # place of residence

data["Value.a4.a2"]=data["Value.a4.a2"].fillna("1") #Ownership of house residence number
data["Value.a5.a1"]=data["Value.a5.a1"].fillna("Own") #Ownership of house


data["Value.a5.a2"]=data["Value.a5.a2"].fillna("1") #type of house  number
data["Value.a6.a1"]=data["Value.a6.a1"].fillna("hut") #typpe of house

data["Value.a7.a1"]=data["Value.a7.a1"].fillna("13") #child-1 age group data["Value.a7.a1"].median() is 13

data["Value.a7.a2"]=data["Value.a7.a2"].fillna(method='ffill') #child-1 age group gender

data["Value.a7.a3"]=data["Value.a7.a3"].fillna("7") #child-1 Education data["Value.a7.a3"].median() is 7


data["Value.a8.a1"]=data["Value.a8.a1"].fillna(data["Value.a8.a1"].mode()) #Type of school
data["Value.a9.a1"]=data["Value.a9.a1"].fillna(data["Value.a9.a1"].mode()) #place of school
```
